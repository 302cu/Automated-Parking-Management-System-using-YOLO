```python
def calculate_iou(box1, box2):
    (x1, y1, w1, h1), (x2, y2, w2, h2) = box1[-4:], box2[-4:]
    box1 = [x1 - w1 / 2, y1 - h1 / 2, w1, h1]
    box2 = [x2 - w2 / 2, y2 - h2 / 2, w2, h2]

    # Calculate intersection coordinates
    x_intersection = max(box1[0], box2[0])
    y_intersection = max(box1[1], box2[1])
    x_overlap = max(0, min(box1[0] + box1[2], box2[0] + box2[2]) - x_intersection)
    y_overlap = max(0, min(box1[1] + box1[3], box2[1] + box2[3]) - y_intersection)

    # Calculate areas
    intersection_area = x_overlap * y_overlap
    box1_area = box1[2] * box1[3]
    box2_area = box2[2] * box2[3]
    union_area = box1_area + box2_area - intersection_area

    # Calculate IoU
    iou = intersection_area / union_area
    return iou

# Read coordinates from text file
def read_coordinates(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        # Assuming each line contains coordinates for one bounding box
        coordinates = [list(map(float, line.strip().split(' '))) for line in lines]
    return coordinates

# Function to calculate IoU between all pairs of bounding boxes
def calculate_iou_matrix(boxes1, boxes2):
    iou_matrix = []
    for box1 in boxes1:
        iou_row = []
        for box2 in boxes2:
            iou_row.append(calculate_iou(box1, box2))
        iou_matrix.append(iou_row)
    return iou_matrix

# File paths
file_path_boxes1 = '/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/labels/parking/42.txt'  # File containing coordinates of boxes 1
file_path_boxes2 = '/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/labels/car/42.txt'  # File containing coordinates of boxes 2

# Read coordinates from files
boxes1_coordinates = read_coordinates(file_path_boxes1)
boxes2_coordinates = read_coordinates(file_path_boxes2)

# Calculate IoU matrix
iou_matrix = calculate_iou_matrix(boxes1_coordinates, boxes2_coordinates)

# Print IoU matrix
for i, row in enumerate(iou_matrix):
    print(f"IoU scores for box {i+1} in parking:")
    for j, iou_score in enumerate(row):
        print(f"  With box {j+1} in car: {iou_score}")
```

```python
import numpy as np
from PIL import Image
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator, Colors

def plot_bounding_boxes(yolo_result, colors=Colors()):
    """Draw bounding boxes from yolo.predict() result"""
    image = yolo_result.orig_img[..., ::-1]
    annotate = Annotator(np.ascontiguousarray(image))
    output_text_path = '/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/labels/car/42.txt'
    classes = yolo_result.boxes.cls.tolist()
    scores = yolo_result.boxes.conf.tolist()
    boxes = yolo_result.boxes.xyxy

    with open(output_text_path, 'w') as f:
        for x, y, w, h in yolo_result.boxes.xywhn.tolist():
            f.write(f"0 {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")

    for box, class_, score in zip(boxes, classes, scores):
        tag = f"{yolo_result.names[class_].title()}: {score:.0%}"
        annotate.box_label(box, tag, colors(class_))

    return Image.fromarray(annotate.result())

model = YOLO('/content/drive/MyDrive/FinalProject/runs/detect/train10/weights/best.pt')

result = model('/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/images/42.png')[0]

plot_bounding_boxes(result)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

parking_path_boxes = '/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/labels/parking/42.txt'  # File containing coordinates of boxes 1
car_path_boxes = '/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/labels/car/42.txt'  # File containing coordinates of boxes 2
image_path ='/content/drive/MyDrive/FinalProject/datasets/cardatasets/compare/images/42.png'

def draw_bounding_boxes(image_path, parking_annotation_path,car_annotation_path):
    # Load the image
    image = cv2.imread(image_path)


    # Load the bounding box coordinates from the annotation file
    parkingbboxes = []
    with open(parking_annotation_path, 'r') as f:
        for line in f:
            values = line.strip().split()
            xc = float(values[1])
            yc = float(values[2])
            w = float(values[3])
            h = float(values[4])
            parkingbboxes.append([xc, yc, w, h])
    carbboxes =[]
    with open(car_annotation_path, 'r') as f:
        for line in f:
            values = line.strip().split()
            xc = float(values[1])
            yc = float(values[2])
            w = float(values[3])
            h = float(values[4])
            carbboxes.append([xc, yc, w, h])


    # Iterate over the bounding boxes and draw them on the image
    for parking_count, bbox in enumerate(parkingbboxes, 0):
        xc = int(bbox[0] * image.shape[1])
        yc = int(bbox[1] * image.shape[0])
        w = int(bbox[2] * image.shape[1])
        h = int(bbox[3] * image.shape[0])
        cv2.rectangle(image, (xc - w // 2, yc - h // 2), (xc + w // 2, yc + h // 2), (0, 255, 0), 2)
        cv2.putText(image, f"Parking {parking_count}", (xc - w // 2, yc - h // 2 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Iterate over the car bounding boxes and draw them on the image
    for car_count, bbox in enumerate(carbboxes, 1):
        xc = int(bbox[0] * image.shape[1])
        yc = int(bbox[1] * image.shape[0])
        w = int(bbox[2] * image.shape[1])
        h = int(bbox[3] * image.shape[0])
        cv2.rectangle(image, (xc - w // 2, yc - h // 2), (xc + w // 2, yc + h // 2), (255, 0, 0), 2)
        cv2.putText(image, f"Car {car_count}", (xc - w // 2, yc - h // 2 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Display the image
    img = image
    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
draw_bounding_boxes(image_path, parking_path_boxes,car_path_boxes)
```
